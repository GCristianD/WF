```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('fraud_dataset.csv', index_col=False)
```

```python
df = df.iloc[:,2:]
```

```python
df.head()
```

```python
df['order_id']  = df['order_id'].str.replace(",","").astype(float)
df['age_of_account'] = df['age_of_account'].str.replace(",","").astype(float)
df['order_total']  = df['order_total'].str.replace(",","").astype(float)
df['total_orders']  = df['total_orders'].str.replace(",","").astype(float)
df['session_duration'] = df['session_duration'].str.replace(",","").astype(float)
df['number_of_pdp_visits_past_hour'] = df['number_of_pdp_visits_past_hour'].str.replace(",","").astype(float)
df['number_of_add_to_cart_events_past_hour'] = df['number_of_add_to_cart_events_past_hour'].str.replace(",","").astype(float)
df['number_of_clicks_performed'] = df['number_of_clicks_performed'].str.replace(",","").astype(float) 
df['number_of_pdp_visits'] = df['number_of_pdp_visits'].str.replace(",","").astype(float) 
df['number_of_keyword_searches_past_hour'] = df['number_of_keyword_searches_past_hour'].str.replace(",","").astype(float) 
df['number_of_pdp_visits_past_day']= df['number_of_pdp_visits_past_day'].str.replace(",","").astype(float)         
df['number_of_add_to_cart_events_past_day']=df['number_of_add_to_cart_events_past_day'].str.replace(",","").astype(float)       
df['number_of_remove_from_cart_events_past_day']=df['number_of_remove_from_cart_events_past_day'].str.replace(",","").astype(float)    
df['number_of_keyword_searches_past_day']=df['number_of_keyword_searches_past_day'].str.replace(",","").astype(float)         
df['number_of_pdp_visits_past_week']=df['number_of_pdp_visits_past_week'].str.replace(",","").astype(float)            
df['number_of_add_to_cart_events_past_week']=df['number_of_add_to_cart_events_past_week'].str.replace(",","").astype(float)   
df['number_of_remove_from_cart_events_past_week']=df['number_of_remove_from_cart_events_past_week'].str.replace(",","").astype(float) 
df['number_of_keyword_searches_past_week']=df['number_of_keyword_searches_past_week'].str.replace(",","").astype(float) 
df['number_of_pdp_visits_past_month']=df['number_of_pdp_visits_past_month'].str.replace(",","").astype(float) 
df['number_of_add_to_cart_events_past_month']=df['number_of_add_to_cart_events_past_month'].str.replace(",","").astype(float) 
df['number_of_remove_from_cart_events_past_month']= df['number_of_remove_from_cart_events_past_month'].str.replace(",","").astype(float) 
df['number_of_keyword_searches_past_month'] = df['number_of_keyword_searches_past_month'].str.replace(",","").astype(float)   
df['billing_and_ip_distance'] = df['billing_and_ip_distance'].str.replace(",","").astype(float)
df['billing_and_ip_distance.1'] = df['billing_and_ip_distance.1'].str.replace(",","").astype(float)
df['customers_with_same_ip_past_day'] = df['customers_with_same_ip_past_day'].str.replace(",","").astype(float)
df['customers_with_same_ip_past_week'] = df['customers_with_same_ip_past_week'].str.replace(",","").astype(float)
df['customers_with_same_ip_past_month'] = df['customers_with_same_ip_past_month'].str.replace(",","").astype(float)
df['customers_with_same_ip_past_six_months'] = df['customers_with_same_ip_past_six_months'].str.replace(",","").astype(float)
```

```python
columns_to_drop = df.columns[df.isnull().mean() > 0.80]
columns_to_drop
```

```python
df = df.drop(columns = columns_to_drop)
```

```python
dummies = pd.get_dummies(df[['currency', 'cvv_code', 'avs_code']])
dummies.head()
```

```python
df = pd.concat([df,dummies], axis=1)
```

```python
dummies = pd.get_dummies(df['ship_class'])
```

```python
N_rows = df.shape[0]
df_shipclass = pd.DataFrame(np.zeros((N_rows, 3)))
df_shipclass.columns = ['freight_truck', 'small_parcel', 'white_glove']
df_shipclass['freight_truck'] = dummies['freight truck,'] + dummies['freight truck,white glove,'] + dummies['small parcel,freight truck,white glove,']+ dummies['small parcel,freight truck,']
df_shipclass['small_parcel'] = dummies['small parcel,freight truck,white glove,'] +dummies['small parcel,white glove,'] + dummies['small parcel,freight truck,'] + dummies['small parcel,'] 
df_shipclass['white_glove'] = dummies['white glove,']+ dummies['small parcel,white glove,']+dummies['small parcel,freight truck,white glove,'] + dummies['freight truck,white glove,']
```

```python
df = pd.concat([df,df_shipclass], axis=1)
df = df.drop(['currency','ship_class','order_id','cvv_code','avs_code'], axis=1)
```

```python
for column in df.columns:
    df.loc[np.isnan(df[column]),column]=df.loc[~np.isnan(df[column]),column].median()
```

```python
df.to_csv('Clean.csv')
```
